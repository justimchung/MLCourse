## The elements of a figure

下圖展示了一張圖裏面的所有的元素。當我們想控制這些元素時 (例如加上 label，legend 時)，這張圖是很好的參考。   
這張圖中，有兩個重要的元素先加以說明。

- Figure: The whole figure. The figure keeps track of all the child Axes, a smattering of ‘special’ artists (titles, figure legends, etc), and the canvas. 整張圖就是一個 figure，打個比方，你在素描時，會有一個畫板，上面放個畫紙，然後在畫紙上面圖色。畫板，畫紙，紙上面的顏色，全部合起來，就是 figure。在 figure 上，我們最常用的功能，就是調整圖的大小。
- Axes: This is what you think of as ‘a plot’, it is the region of the image with the data space (marked as the inner blue box). A given figure can contain many Axes, but a given Axes object can only be in one Figure. Axes 就是畫紙。大部份 seaborn 繪圖的功能在畫完圖後，都會回傳一個 axes。我們可以透過 axes 控制圖的標題，圖的 x/y 軸標籤等。

## import packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Numpy and seaborn

In the following, we show how to use seaborn to draw data stored in the numpy array.

雖然 seaborn 在設計上是較傾向於 pandas，但它依然可以和 numpy 合作。   
底下我們示範如何使用 seaborn 來繪出 numpy 中的資料。

```python
#產生 x 軸的資料
x = np.linspace(1.0, 10.0, num=40)
#產生 y 軸的資料。在 [10, 50) 區間中，隨機產生 40 個點
y = np.random.randint(10, 50, size=40)
# seaborn 
sns.scatterplot(x, y)
plt.ylabel('y values')
plt.xlabel('x values')
plt.title('This is the title of the graph')
```

### Add legend to the data

替資料加上標籤 (legend)

```python
#產生 x 軸的資料
x = np.linspace(1.0, 10.0, num=40)
#產生 2 個 y 軸的資料。在 [10, 50) 區間中，隨機產生 40 個點
y1 = np.random.randint(10, 50, size=40)
y2 = np.random.normal(30, 2, size=40) # normal distribution with mean=30 and std = 2
# seaborn, label 指令會幫資料加上標籤 (legend)
sns.scatterplot(x, y1, label='Uniform')
sns.scatterplot(x, y2, label='Gaussian')
plt.ylabel('y values')
plt.xlabel('x values')
plt.title('This is the title of the graph')
```

### Control the size of the figure

調整圖的大小

```python
#使用 figure 來調整圖的大小
plt.figure(figsize=(8,7))
#產生 x 軸的資料
x = np.linspace(1.0, 10.0, num=40)
#產生 2 個 y 軸的資料。在 [10, 50) 區間中，隨機產生 40 個點
y1 = np.random.randint(10, 50, size=40)
y2 = np.random.normal(30, 2, size=40) # normal distribution with mean=30 and std = 2
# seaborn, label 指令會幫資料加上標籤 (legend)
sns.scatterplot(x, y1, label='Uniform')
sns.scatterplot(x, y2, label='Gaussian')
plt.ylabel('y values')
plt.xlabel('x values')
plt.title('This is the title of the graph')
```

### Subplot

subplot adds a subplot to the current figure.

> subplot(nrows, ncols, index)  
> nrows: 子圖中有幾個 rows (number of rows in the sub plot)  
> ncols: 子圖中有幾個 columns (number of columns in the sub plot)  
> index: 目前子圖的索引值 (由 1 開始) (The index of the current sub plot)    

The following figure shows the subplots using ``plt.subplot(2,3,i)``.  
Please refer to [here](https://ithelp.ithome.com.tw/articles/10201670) for the detailed explaination.

![](https://ithelp.ithome.com.tw/upload/images/20181016/20111390EGdCZj8RDK.png)

```python
#產生資料
#產生 x 軸的資料
x = np.linspace(1.0, 10.0, num=40)
#產生 2 個 y 軸的資料。在 [10, 50) 區間中，隨機產生 40 個點
y1 = np.random.randint(10, 50, size=40)
y2 = np.random.normal(30, 2, size=40) # normal distribution with mean=30 and std = 2
#繪圖
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.scatterplot(x, y1)
plt.ylim((10, 50))
plt.xlabel('X values')
plt.ylabel('Y values (uniform)')
plt.title('Uniform')

plt.subplot(1,2,2)
sns.scatterplot(x, y2)
plt.ylim((10,50)) #
plt.xlabel('X values')
plt.ylabel('Y values (Gaussian)')
plt.title('Gaussian')
```

## Draw data in Pandas

In the following, we show how to use seaborn to visualize data points in the Pandas dataframe.

### Load the dataset

```python
df = pd.read_csv('gapminder.csv')
df.head()
```

## Scatter Plot

Draw the scatter Plot for gdpPercap vs. LifeExp

```python
df1 = df.loc[df.country.isin(['Taiwan','Japan','China']), :]
df1.head(5)
```

```python
sns.scatterplot(x=df1.gdpPercap, 
                y=df1.lifeExp,
                hue = df1.country)
plt.title('GDP VS. Life Expectancy')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy (year)')9
```

## Line Plot

```python
sns.lineplot(x = df1.year, 
             y = df1.lifeExp, 
             hue = df1.country)
```

## Bar chart

A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars. Bar plots include 0 in the quantitative axis range, and they are a good choice when 0 is a meaningful value for the quantitative variable, and you want to make comparisons against it.

Barplot 會使用平均值繪出資料的分佈。  
每個 bar 上的直線代表 95% 的信賴區間 (confidence interval; CI)  
你可以使用 ``ci=None`` 這個選項來取消信賴區間。

[Barplot documentation](http://seaborn.pydata.org/generated/seaborn.barplot.html?highlight=barplot#seaborn.barplot)

```python
df2007 = df.loc[(df.year == 2007), :]
df2007.head()
```

```python
sns.barplot(x = df2007.continent, 
            y = df2007.lifeExp)
```

由底下的圖可以看出來，每一洲的 lifeexp 的均值都隨著年代的前進，而增長。

```python
plt.figure(figsize=(15, 8))
sns.barplot(x = df.continent, y = df.lifeExp, hue=df.year, ci=None)
```

## Countplot

Show the counts of observations in each categorical bin using bars.

[Documentation of countplot](http://seaborn.pydata.org/generated/seaborn.countplot.html?highlight=countplot#seaborn.countplot)

Countplot 則是繪出每一種類別資料的個數。

```python
print('The number of countries in each continent.')
print(df2007.continent.value_counts())
sns.countplot(x = df2007.continent)
plt.ylabel('Number of countries');
```

## Box Plot

A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

[Documentation of Boxplot](http://seaborn.pydata.org/generated/seaborn.boxplot.html?highlight=boxplot#seaborn.boxplot)

```python
sns.boxplot(x = df2007.continent, 
            y = df2007.gdpPercap)
```

## Violin plot

Draw a combination of boxplot and kernel density estimate.

A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

[Documentation of violinplot](http://seaborn.pydata.org/generated/seaborn.violinplot.html?highlight=violinplot#seaborn.violinplot)

Villin plot 是 boxplot 還有 kernel density (資料分佈圖) 的混合體

```python
plt.figure(figsize=(10, 8))
sns.violinplot(x = df2007.continent,
              y=df2007.gdpPercap)
```
